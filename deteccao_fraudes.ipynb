## Padronização dos dados

### Estado Civil 

SOLTEIRO(A)	= 1  <br> 
CASADO (A)	= 2  <br> 
DIVORCIADO	= 3 <br> 
OUTRO = 4 <br> 
VIÚVO(A) = 5 <br> 
UNIÃO ESTAVEL = 6 <br> 
SEPARADO JUDICIALMENTE = 7 <br> 
NENHUM = 8
### Sexo

Feminino = 1  <br> 
Masculino = 2

### Escolaridade

nan = 0 <br>
Nenhum = 1 <br>
Ensino Médio = 2 <br>
Ensino Superior = 3 <br>
Ensino Fundamental = 4 <br>
Pós Graduação / Mestrado / Doutorado = 5

### Possui patrimonio
Sim = 1 <br>
Não = 2

### Possível Fraude

Sim = 1 <br>
Não = 2

```python
base_tratada = base_original
```

```python
#Criando um dificionario que mapeia a alteração dos dados 

mapeamento_estado_civil = {
    'SOLTEIRO(A)': 1,
    'CASADO (A)': 2,
    'DIVORCIADO': 3,
    'OUTRO': 4,
    'VIÚVO(A)': 5,
    'UNIÃO ESTAVEL': 6,
    'SEPARADO JUDICIALMENTE': 7,
    'NENHUM': 8
}

mapeamento_sexo = {
    'F': 1,
    'M': 2
}

mapeamento_escolaridade = {
    'nan': 0,
    'Nenhum': 1,
    'Ensino Médio': 2,
    'Ensino Superior': 3,
    'Ensino Fundamental': 4,
    'Pós Graduação / Mestrado / Doutorado': 5
}

mapeamento_patrimonio = {
    'S': 1,
    'N': 2
}

mapeamento_possivel_fraude = {
    'Sim': 1,
    'Nao': 2
}
```

```python
base_tratada['Estado_Civil'] = base_tratada['Estado_Civil'].replace(mapeamento_estado_civil)
base_tratada['Sexo'] = base_tratada['Sexo'].replace(mapeamento_sexo)
base_tratada['Escolaridade'] = base_tratada['Escolaridade'].replace(mapeamento_escolaridade)
base_tratada['Possui_Patrimonio'] = base_tratada['Possui_Patrimonio'].replace(mapeamento_patrimonio)
base_tratada['Possivel_Fraude'] = base_tratada['Possivel_Fraude'].replace(mapeamento_possivel_fraude)
```

```python
#Exibição da base tratada
base_tratada[['Estado_Civil', 'Sexo', 'Escolaridade','Possui_Patrimonio', 'Possivel_Fraude']]
```

## Tratando os dados núlos

Como agora estão na escala númerica, é possível utilizar a média e mediana para preencher os valores nulos da escolaridade 


```python
#Tratando os núlos da escolaridade
media_escolaridade = base_tratada['Escolaridade'].mean()
print(media_escolaridade)
```

```python
base_tratada['Escolaridade'].fillna(media_escolaridade, inplace=True)
```

```python
#A coluna veio com o tipo float, para padrozinar estou deixando tudo como inteiro
base_tratada['Escolaridade'] = base_tratada['Escolaridade'].astype(int)
```
